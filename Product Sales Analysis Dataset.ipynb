```python
#drop the last 2 columns as they dont have any data
df.drop(["Status", "unnamed1"], axis = 1, inplace = True)
```

```python
#Also yopu can see the amount of non null value in Amount column is low as compared to others, so lets go with nul value:
df.isnull()
#this instruction will give you the whole table, which is time consuming for data reading, thus use df.isnull().sum()
```

```python
#to get the count of null values in a column, we use this instruction
df.isnull().sum()
```

```python
# to drop the null values
df.dropna(inplace = True)

# "df = df.dropna()" is same as "df.dropna(inplace = True)"
```

```python
# df.shape
# OR
df.isnull().sum()
```

```python
#changing datatype of amount from float to int
df["Amount"] = df["Amount"].astype("int")
df.dtypes
```

```python
#changing the name of column Product_Category to Categories
df.rename(columns = {"Product_Category" : "Categories"},inplace = True)
```

```python
ax = sns.countplot(x = "Gender", data = df)
#t show values of bars on the top of bars
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#sales amount by gender
sale_gen = df.groupby(["Gender"], as_index  = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
sns.barplot(x="Gender",y="Amount",data= sale_gen)
```

```python
#try without hue
sns.countplot(data = df, x = "Age Group")
```

```python
#try with hue
sns.countplot(data = df, x = "Age Group",hue = "Gender")
```

```python
#sales amount by Age groups
sale_gen = df.groupby(["Age Group"], as_index  = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
sns.barplot(x="Age Group",y="Amount",data= sale_gen)
```

```python
sales_state = df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by = "Orders", ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = "State", y = "Orders")
```

```python
sales_state = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending = False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = "State", y = "Amount")
```

```python
ms = sns.countplot(x = "Marital_Status", data = df)
sns.set(rc={'figure.figsize':(5,5)})
#t show values of bars on the top of bars
for bars in ms.containers:
    ms.bar_label(bars)
```
